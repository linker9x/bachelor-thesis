# NN - distance plot

```python
for name in names:
    df = pd.read_csv('../data/' + name + '_full.csv')
    
    # features and target var
    X = df.drop('class', axis=1)
    y = df['class']
    print(X.shape)
    print(y.shape)
    
    # KNN
    k_range = range(1,6)
    for k in k_range:
        nn = NearestNeighbors(n_neighbors=k)
        nn.fit(X)
        distances, indices = nn.kneighbors(X)
        distanceDec = sorted(distances[:,k-1], reverse=True)
        
        plt.figure(figsize=(10,8))
        plt.plot(list(range(1,len(X)+1)), distanceDec)
        plt.grid(True)
```

# DBSCAN

```python
for name in names:
    df = pd.read_csv('../data/' + name + '_full.csv')
    
    # features and target var
    X = df.drop('class', axis=1)
    y = df['class']
    
    eps_range = range(160, 220, 1)
    result = []
    for eps in eps_range:
        db = DBSCAN(eps=eps)
        db.fit(X)
        labels = db.labels_
        result.append(len(np.unique(labels)))
    print(result)
```
